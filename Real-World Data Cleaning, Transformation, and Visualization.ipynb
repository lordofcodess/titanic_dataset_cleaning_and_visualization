```python
import pandas as pd
import numpy as np  
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset('titanic')
data.info()
data.head()


```

```python
data.isnull().sum()
```

```python
# Drop the deck column, it won't be used.
# Remove null rows 
```

```python
data.drop('deck', axis = 1, inplace = True)
```

```python
data.info()
```

```python
data['age'].unique()
```

```python
data['age'].fillna(data['age'].mean(), inplace=True)
```

```python
data.dropna(inplace = True)
```

```python
data.info()
```

```python
data.head(2)
```

```python
data['family_size'] = data['sibsp'] + data['parch'] + 1
```

```python
data.head()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=data, x='age', kde=True)
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=data, x='survived', y='fare')
plt.title("Comparison of Fares by Survival")
plt.xlabel("Survived")
plt.ylabel("Fare")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
columns = ['survived', 'age', 'fare', 'pclass', 'sibsp', 'parch']

# Create a pairplot
sns.pairplot(data=data[columns], hue='survived', markers=['o', 's'], diag_kind='hist')
plt.suptitle("Pairwise Relationship with Survival")
plt.show()
```
